<h1>Student grades on a chemistry exam were: 77, 78, 76, 81, 86, 51, 79, 82, 84, 99
Construct a stem-and-leaf plot of the data.
</h1>

```python
import stemgraphic
import matplotlib.pyplot as plt
data=[77, 78, 76, 81, 86, 51, 79, 82, 84, 99]
stem=[7,7,7,7,8,8,8,8,5,9]
plt.ylabel('Data')   
plt.xlabel('stems')  
  
plt.stem(stem, data)
stemgraphic.stem_graphic(data, scale = 10)
```
